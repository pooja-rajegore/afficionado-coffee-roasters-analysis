```python
import pandas as pd
```

```python
df = pd.read_excel("Data/Raw_Afficionado_Coffee_Roasters.xlsx")
```

```python
df.head()
```

```python
#Check Missing Values
df.isnull().sum()
```

```python
#Check Duplicate Records
df.duplicated().sum()
```

```python
#Verify Data Types
df.info()
```

```python
#Convert Time Column
df['transaction_time'] = pd.to_datetime(
    df['transaction_time'],
    format='%H:%M:%S'
).dt.time
```

```python
#Hours Column
df['hour'] = df['transaction_time'].dt.hour
```

```python
#Count product_category
df['product_category'].value_counts()
```

```python
#Check Negative or Zero Values
df[df['transaction_qty'] <= 0]
df[df['unit_price'] <= 0]
```

```python
#Remove Invalid Records
df = df[df['transaction_qty'] > 0]
df = df[df['unit_price'] > 0]
```

```python
#Create Revenue Column
df['revenue'] = df['transaction_qty'] * df['unit_price']
```

```python
#Sales Session
def session(hour):
    if hour < 12:
        return 'Morning'
    elif hour < 17:
        return 'Afternoon'
    else:
        return 'Evening'

df['session'] = df['hour'].apply(session)
```

```python
#Save cleaned data file
df.to_excel("Cleaned_Afficionado_Coffee_Data.xlsx", index=False)
```
