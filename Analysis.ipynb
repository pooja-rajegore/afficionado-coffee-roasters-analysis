```python
import pandas as pd
```

```python
#Load Cleaned Data
df=pd.read_excel('data/Cleaned_Afficionado_Coffee_Data.xlsx')
```

Revenue Computation

```python
#Aggregate Revenue by Product
product_revenue = (
    df.groupby('product_detail')['revenue']
      .sum()
      .reset_index()
      .sort_values(by='revenue', ascending=False)
)

print(product_revenue.head())
```

```python
#Aggregate Revenue by Product Type
product_type_revenue = (
    df.groupby('product_type')['revenue']
      .sum()
      .reset_index()
      .sort_values(by='revenue', ascending=False)
)

print(product_type_revenue)
```

```python
#Aggregate Revenue by Product Category
category_revenue = (
    df.groupby('product_category')['revenue']
      .sum()
      .reset_index()
      .sort_values(by='revenue', ascending=False)
)

print(category_revenue)
```

```python
with pd.ExcelWriter("Revenue_Analysis.xlsx") as writer:
    product_revenue.to_excel(
        writer,
        sheet_name="Product Revenue",
        index=False
    )

    product_type_revenue.to_excel(
        writer,
        sheet_name="Product Type Revenue",
        index=False
    )

    category_revenue.to_excel(
        writer,
        sheet_name="Category Revenue",
        index=False
    )
```

Revenue Contribution Analysis

```python
#Total Revenue per Product
product_revenue = (
    df.groupby('product_detail')['revenue']
      .sum()
      .reset_index()
      .sort_values(by='revenue', ascending=False)
)

product_revenue.head()
```

```python
#Revenue Share (%) of Each Product
total_revenue = product_revenue['revenue'].sum()

product_revenue['revenue_share_pct'] = (
    product_revenue['revenue'] /
    total_revenue
) * 100

product_revenue.head()
```

```python
#Comparison Between Volume Rank and Revenue Rank
#Create Sales Volume Table
product_volume = (
    df.groupby('product_detail')['transaction_qty']
      .sum()
      .reset_index()
)
```

```python
#Add Volume Rank
product_volume['volume_rank'] = (
    product_volume['transaction_qty']
    .rank(method='dense', ascending=False)
    .astype(int)
)
```

```python
#Add Revenue Rank
product_revenue['revenue_rank'] = (
    product_revenue['revenue']
    .rank(method='dense', ascending=False)
    .astype(int)
)
product_revenue['revenue_rank'] = (
    product_revenue['revenue']
    .rank(method='dense', ascending=False)
    .astype(int)
)
```

```python
#Merge Both Tables
comparison = pd.merge(
    product_volume,
    product_revenue[
        ['product_detail',
         'revenue',
         'revenue_share_pct',
         'revenue_rank']
    ],
    on='product_detail'
)
comparison.head()
```

```python
# save Revenue Contribution Analysis in Excel sheet
with pd.ExcelWriter(
    "Revenue_Analysis.xlsx",
    engine="openpyxl",
    mode="a",
    if_sheet_exists="replace"
) as writer:

    comparison.to_excel(
        writer,
        sheet_name="Revenue Contribution",
        index=False
    )
```

```python
#Product-type contribution within each category
type_contribution = (
    df.groupby(
        ['product_category', 'product_type']
    )['revenue']
      .sum()
      .reset_index()
)
type_contribution
```

Revenue Concentration & Menu Balance

```python
#Pareto (80/20) Analysis
total_revenue = product_revenue['revenue'].sum()

product_revenue['revenue_share_pct'] = (
    product_revenue['revenue'] /
    total_revenue
) * 100
```

```python
#Calculate Cumulative Revenue %
product_revenue['cumulative_pct'] = (
    product_revenue['revenue_share_pct']
    .cumsum()
)
product_revenue.head()
```

```python
#Find Products Generating 80% Revenue
pareto_products = product_revenue[
    product_revenue['cumulative_pct'] <= 80
]

pareto_products
```

```python
#Count Products Driving 80% Revenue
num_products = len(pareto_products)

total_products = len(product_revenue)

print(
    f"{num_products} out of {total_products} products generate 80% of revenue"
)
```

```python
#Identify Revenue Anchors
revenue_anchors = product_revenue.head(10)

revenue_anchors
```

```python
#Identify Long-Tail Products
long_tail_products = product_revenue.tail(20)

long_tail_products
```

```python
#Revenue Contribution of Long Tail
long_tail_share = (
    long_tail_products['revenue'].sum()
    / total_revenue
) * 100

print(long_tail_share)
```

```python
#Menu Diversification Risk
#Top 5 Product Concentration
top5_revenue = (
    product_revenue.head(5)['revenue']
    .sum()
)

top5_ratio = (
    top5_revenue /
    total_revenue
) * 100

print(top5_ratio)
```

```python
#Save Result
with pd.ExcelWriter(
    "Revenue_Analysis.xlsx",
    engine="openpyxl",
    mode="a",
    if_sheet_exists="replace"
) as writer:

    product_revenue.to_excel(
        writer,
        sheet_name="Pareto Analysis",
        index=False
    )

    revenue_anchors.to_excel(
        writer,
        sheet_name="Revenue Anchors",
        index=False
    )

    long_tail_products.to_excel(
        writer,
        sheet_name="Long Tail Products",
        index=False
    )
```

KPI Calculation

```python
#Product Revenue Contribution (%)
product_revenue = (
    df.groupby('product_detail')['revenue']
      .sum()
      .reset_index()
)

total_revenue = product_revenue['revenue'].sum()

product_revenue['revenue_contribution_pct'] = (
    product_revenue['revenue'] /
    total_revenue
) * 100

product_revenue.sort_values(
    by='revenue_contribution_pct',
    ascending=False
).head()
```

```python
#Product Sales Volume
product_volume = (
    df.groupby('product_detail')['transaction_qty']
      .sum()
      .reset_index()
      .sort_values(
          by='transaction_qty',
          ascending=False
      )
)

product_volume.head()
```

```python
#Category Revenue Share
category_revenue = (
    df.groupby('product_category')['revenue']
      .sum()
      .reset_index()
)

total_revenue = category_revenue['revenue'].sum()

category_revenue['category_share_pct'] = (
    category_revenue['revenue']
    / total_revenue
) * 100

category_revenue
```

```python
#Revenue Concentration Ratio
product_revenue = (
    df.groupby('product_detail')['revenue']
      .sum()
      .sort_values(ascending=False)
)

top5_revenue = product_revenue.head(5).sum()

total_revenue = product_revenue.sum()

revenue_concentration_ratio = (
    top5_revenue /
    total_revenue
) * 100

print(revenue_concentration_ratio)
```

```python
#Product Efficiency Score
total_revenue = df['revenue'].sum()

sku_count = df['product_detail'].nunique()

product_efficiency_score = (
    total_revenue /
    sku_count
)

print(product_efficiency_score)
```

```python
#Product-Level Efficiency
product_efficiency = (
    df.groupby('product_detail')['revenue']
      .sum()
      .reset_index()
      .sort_values(
          by='revenue',
          ascending=False
      )
)
```
